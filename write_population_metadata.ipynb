```python
import csv
from cyvcf2 import VCF
demo_names = ["const", "bottleneck", "exp", "split", "step"]
for i in range(1,2):
    vcf_path = f'example_data/diploid_sim/{demo_names[i]}/{demo_names[i]}.vcf.gz'
    meta_path = f'example_data/diploid_sim/{demo_names[i]}/{demo_names[i]}.meta.csv'
    
    samples = VCF(vcf_path).samples  # guaranteed to match VCF column order
    
    with open(meta_path, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["population", "name"])
        for name in samples:
            parts = name.split("_")
            if i == 4:
                pop =  "_".join(parts[1:3])
            else:
                pop = parts[1] # -> "D0"
            writer.writerow([pop, name])
```
